```python
#open metadata file
metadata = pd.read_csv("/mnt/raid1/armen/Schinos/META-ANALYSIS/ALL/Fungi/RNA_metadata_ALL.tab", sep= "\t")
#metadata.head()
sample_names= metadata['External_ID'].tolist()
#sample_names
```

```python
# Filter the DataFrame based on the 'Condition' column
healthy_df = metadata.loc[metadata['Condition'] == 'Healthy']
periodontitis_df = metadata.loc[metadata['Condition'] == 'Periodontitis']

# Convert the 'External ID' column to a list for each filtered DataFrame
healthy_ids = healthy_df['External_ID'].tolist()
periodontitis_ids = periodontitis_df['External_ID'].tolist()

# Display the lists of 'External ID' values
print('Healthy IDs:', healthy_ids)
print('Periodontitis IDs:', periodontitis_ids)
```

```python
samfiles=[]
for sample in sample_names:
    samfile=path + sample +'/'+ sample +'_vir_mapping.sam'
    samfiles.append(samfile)
samfiles[0]
```

```python
### to be able to run bash command loop in python
import subprocess
# Loop 
for samfile in samfiles:
    # Define the bash script to run for each element
    samtools_script =f'samtools view -F 4 {samfile} | cut -f 3 | sort | uniq -c | sort -k1,1rn > /mnt/raid1/armen/Schinos/META-ANALYSIS/ALL/Fungi/VirulenceGenes/{os.path.basename(samfile)}_sorted_vir_genes.txt'
    #This command filters the SAM file for reads that are mapped(-F 4) (if we want we can use -f 2 to filter the paired), 
    #and then extracts the reference name field (column 3) using the cut command. The output will be a 
    #list of reference names, one per line, for each mapped read. 
    #Finally, you can use the uniq -c command to count and then sort based on the counts of each reference name

    # Use subprocess to run the bash script
    subprocess.run(samtools_script, shell=True)    
    
#make a list of all files (sorted_vir_genes) in the directory VirulenceGenes that we created  
# Get a list of all files and directories in the directory
all_files = os.listdir(path)
# Filter out the directories and hidden files to get only the file names
file_names = [file for file in all_files if os.path.isfile(os.path.join(path, file)) and not file.startswith('.')]
# Print the list of file names
#print(file_names)   
len(file_names)
```

```python
#Save the names of each virulence factor genes from the VFDB in a list. Have a split version of it
virulence_genes_fasta='/mnt/raid1/armen/Schinos/META-ANALYSIS/Databases/VFDB/VFDB_nt.fa'

fasta_file = open(virulence_genes_fasta, "r")
reference_names = []
reference_split=[]
for line in fasta_file:
    if line.startswith(">"):
        # Extract the reference name from the header line
        reference_name = line[1:].strip()
        reference_names.append(reference_name)
        reference_split.append(reference_name.split())
fasta_file.close()

print('The total number of virulence factor genes in the VFDB database is: ' + str(len(reference_names)))
print(reference_names[0])
#print(len(reference_split))
#print(reference_split[1])
```

```python
#save the full name of the virulence gene found in the database
with open('VirGenes_fullnames.txt', 'w') as file:
    # Write each value on a new line separated by tabs
    for value in reference_names:
        file.write(value + '\n')
```

```python
## For each sample, make a csv file with the Virulence Factor reference(NCBI gene accession),
#Gene name, Encoded protein,Function, Origin and Counts


#for each sample
for f in file_names:
    #open file
    fl = open(path + f, "r")
    read = fl.readlines()
    fl.close()
    vir_genes=[]
    #strip each line and split in 2 parts(counts, reference) and store in the list 'lines'
    vir_genes = [l.strip().split() for l in read]
    #vir_genes[0]
    
    #a dictionary of identified virulence factor genes and their counts
    vir_dict = {item[1]: item[0] for item in vir_genes}
    #print(vir_dict)
    
    #a dictionary with the full reference name and their counts
    items = list(vir_dict.items())
    full_dict=dict()
    for i in range(len(reference_split)):
        for j in range(len(items)):
            if reference_split[i][0]==items[j][0]:
                full_dict[reference_names[i]]=items[j][1]

    #Make a dictionary with the virulence factor genes sorted (decreasing) by the number of their counts
    # Convert the values to integers
    full_dict = {k: int(v) for k, v in full_dict.items()}
    # Sort the dictionary based on the values in decreasing order
    full_dict_sorted = dict(sorted(full_dict.items(), key=lambda item: item[1], reverse=True))
    # Print the sorted dictionary
    #print(len(full_dict_sorted))
    #print(full_dict_sorted)
    
    #the patterns to check in the full reference name 
    pattern_gene = r'\([^()]+\).*?\([^()]+\)\s+(.+?)\s*\['
    # In the above pattern, we first match any sequence of characters that is enclosed in parentheses '\([^()]+\)'. 
    # This matches the first set of parentheses e.g. (gb|WP_006154814).
    # We then match any sequence of characters non-greedily, i.e., matching the shortest possible sequence of characters,
    # using '.*?'. This matches the characters between the first set of parentheses and the second set of parentheses.
    # Next, we match the second set of parentheses \([^()]+\) and one or more spaces \s+, 
    # followed by the substring you want (.+?), which is the shortest possible sequence of one or more characters ending 
    # with a space. Finally, we match zero or more spaces \s* followed by the character '['.

    pattern_function = r'(?<=\s-\s)[\w\s/]+(?=\()'
    #This pattern uses a positive lookahead (?=\() to match all the words after '-' and before the next '('. 
    #The [\w\s/]+ matches one or more word characters and spaces (including '/'). 
    #The pattern is surrounded by the lookbehind (?<=\s-\s) and lookahead (?=\() assertions.

    pattern_species = r'\[([^\[\]]+)\][^\[\]]*\['
    # In this pattern, we use square brackets \[ and \] to match literal opening and closing brackets, respectively. The 
    # ([^\[\]]+) pattern matches one or more characters that are not brackets and captures them in a group.
    # The [^\[\]]*\[ pattern matches any characters that are not brackets
    # and followed by an opening bracket. This ensures that we are selecting the second pair of brackets in the string.

    # Convert the dictionary to a list of tuples
    table_data = [(key.split()[0][:10] + key.split()[0][13:], key.split()[1].strip('()'), \
                   re.search(pattern_gene, key).group(1) if re.search(pattern_gene, key) else '', \
                   re.search(pattern_function, key).group(0) if re.search(pattern_function, key) else '', \
                   re.search(pattern_species, key).group(1) if re.search(pattern_species, key) else '', value)\
                   for key, value in full_dict_sorted.items()]

    # Print the table
    #print(tabulate(table_data, headers=['Virulence Factor reference(NCBI gene accession)','Gene name', 'Gene',\
    #                                    'Function', 'Species', 'Counts']))
    
    # For the proper naming of the files
    # Find the starting index of the desired substring
    start_index = f.find("_vir_mapping.sam_sorted_vir_genes.txt")
    # Extract the desired substring
    desired_string = f[:start_index]
    # Create the file name using the desired substring
    file_name = desired_string + "_results"

    
    # Save the table as a tab-delimited file
    with open(file_name, "w", newline="") as file:
        writer = csv.writer(file, delimiter="\t")
        writer.writerow(['Virulence Factor reference(NCBI gene accession)','Gene name', 'Encoded protein', 'Function',\
                     'Origin', 'Counts'])
        for row in table_data:
            writer.writerow(row)
```

```python
#Here we extract the 'Virulence Factor reference(NCBI gene accession)' and 'Counts' columns from 
# each dataframe. We then rename the 'Counts' column with the respective dataframe name. Finally, we merge the 
# dataframes based on the 'Virulence Factor reference(NCBI gene accession)' column and set it as the row index 
# in the merged dataframe.

# Create an empty list to store the DataFrames
df_healthy = []
df_peri=[]

# Loop through the healthy and periodontitis samples and append to a list of dataframes
for sample in healthy_ids:
    df = pd.read_csv(sample + "_results", sep='\t')
    df_healthy.append(df)

for sample in periodontitis_ids:
    df = pd.read_csv(sample + "_results", sep='\t')
    df_peri.append(df)    


# Create an empty dataframe to store the merged data
merged_healthy = pd.DataFrame()
merged_peri = pd.DataFrame()


# Iterate over the dataframes in the list
#helathy
for i, df in enumerate(df_healthy):
    # Extract the 'Virulence Factor reference(NCBI gene accession)' and 'Counts' columns
    df_counts = df[['Virulence Factor reference(NCBI gene accession)', 'Counts']]
    
    # Rename the 'Counts' column with the dataframe name
    df_counts.rename(columns={'Counts': healthy_ids[i]}, inplace=True)
    
    # Merge the current dataframe with the merged dataframe
    if merged_healthy.empty:
        merged_healthy = df_counts
    else:
        merged_healthy = merged_healthy.merge(df_counts, how='outer', on='Virulence Factor reference(NCBI gene accession)')

#Periodontitis
for i, df in enumerate(df_peri):
    # Extract the 'Virulence Factor reference(NCBI gene accession)' and 'Counts' columns
    df_counts = df[['Virulence Factor reference(NCBI gene accession)', 'Counts']]
    
    # Rename the 'Counts' column with the dataframe name
    df_counts.rename(columns={'Counts': periodontitis_ids[i]}, inplace=True)
    
    # Merge the current dataframe with the merged dataframe
    if merged_peri.empty:
        merged_peri = df_counts
    else:
        merged_peri = merged_peri.merge(df_counts, how='outer', on='Virulence Factor reference(NCBI gene accession)')

        
        
# Set the 'Virulence Factor reference(NCBI gene accession)' as the row index
#healthy
merged_healthy.set_index('Virulence Factor reference(NCBI gene accession)', inplace=True)
#periodontitis
merged_peri.set_index('Virulence Factor reference(NCBI gene accession)', inplace=True)

#merge healthy and periodontitis dataframes together
merged_ALL = pd.merge(merged_healthy, merged_peri, on='Virulence Factor reference(NCBI gene accession)')
# Fill all NaN values with zeros
df_filled = merged_ALL.fillna(0)

# Save the merged dataframes as a tab-delimited CSV file
merged_healthy.to_csv('VirGenes_healthy.csv', sep='\t', index=True)
merged_peri.to_csv('VirGenes_peri.csv', sep='\t', index=True)
df_filled.to_csv('VirGenes_ALL.csv', sep='\t')


# Print the merged dataframe
#print(merged_healthy)
#print(merged_peri)
```
